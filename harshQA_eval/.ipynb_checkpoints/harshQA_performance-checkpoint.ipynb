### 2) Performance de ranking:  rang de la première réponse 

```python
r1_dummies.Model.unique()
```

```python
r1_dummies[(r1_dummies.Model=='BERT TUNED Skmeans' ) & (r1_dummies.Q_id==8) & (r1_dummies.Company=='Sanofi')]
```

```python
def create_new_rank(row):
    i=row.name
    same_model=r1_dummies[(r1_dummies['Q_id']==row['Q_id'])&(r1_dummies['Company']==row['Company'])&(r1_dummies['Model']==row['Model'])&(r1_dummies['Domain']==row['Domain'])& (r1_dummies['Annoted_as_candidate']==1)]
    other_model=r1_dummies[(r1_dummies['Q_id']==row['Q_id'])&(r1_dummies['Company']==row['Company'])&(r1_dummies['Model']!=row['Model'])&(r1_dummies['Domain']==row['Domain'])& (r1_dummies['Annoted_as_candidate']==1)]
    if len(same_model)==0:
        if len(other_model)==0:
            row['rank_first_candidate']=0
        else:
            row['rank_first_candidate']=100
    else:
        rank=same_model.iloc[0]['Rank']
        row['rank_first_candidate']=int(rank)
    return row

r1_dummies=r1_dummies.apply(create_new_rank,axis=1)
```

```python
def create_new_rank2(row):
    i=row.name
    same_model=r1_dummies[(r1_dummies['Q_id']==row['Q_id'])&(r1_dummies['Company']==row['Company'])&(r1_dummies['Model']==row['Model'])&(r1_dummies['Domain']==row['Domain'])& (r1_dummies['Annotation_answer']==1)]
    other_model=r1_dummies[(r1_dummies['Q_id']==row['Q_id'])&(r1_dummies['Company']==row['Company'])&(r1_dummies['Model']!=row['Model'])&(r1_dummies['Domain']==row['Domain'])& (r1_dummies['Annotation_answer']==1)]
    if len(same_model)==0:
        if len(other_model)==0:
            row['rank_first_answer']=0
        else:
            row['rank_first_answer']=100
        
    else:
        rank=same_model.iloc[0]['Rank']
        row['rank_first_answer']=(int(rank))
    return row

r1_dummies=r1_dummies.apply(create_new_rank2,axis=1)
```

```python
basis_ranker=r1_dummies[r1_dummies['rank_first_candidate']!=0]
basis_ranker_answer=r1_dummies[r1_dummies['rank_first_answer']!=0]
```

```python
percentile_10 = lambda x: np.percentile(x, 10) 
percentile_10.__name__ = 'percentile_10'
percentile_25 = lambda x: np.percentile(x, 25)
percentile_25.__name__ = 'percentile_25'
percentile_75 = lambda x: np.percentile(x, 75)
percentile_90 = lambda x: np.percentile(x, 90)
percentile_90.__name__ = 'percentile_90'
percentile_75.__name__ = 'percentile_75'
def f_ranker(x):
    dic={}
    dic['median']=x.median()
    dic['mean']=x.mean()
    dic['q1']=np.percentile(x,25)
    dic['q3']=np.percentile(x,75)
    return pd.Series(dic,index=dic.keys())
ranker=basis_ranker.groupby(['Model'])['rank_first_candidate'].agg({"rank":[np.mean,percentile_10,percentile_25,np.median,percentile_75,percentile_90]})
ranker
```

### 3) MAP Mean Average Precision:

```python
def func_AP(row):
    i=row.name
    
    m,qid,c=row.Model,row.Q_id,row.Company
    same=r1_dummies[(r1_dummies.Model==m )& (r1_dummies.Company==c) & (r1_dummies.Q_id==qid)]
    count=0
    count_cum=0

    
    for k in range(5):
        label=same.iloc[k]['Annoted_as_candidate']
        if label==1:
            count+=1
            count_cum+=count/(k+1)
    if count==0:
        return 0
    else:
        return count_cum/count
    

r1_dummies['AP']=r1_dummies.apply(func_AP,axis=1)
```

```python
r1new=r1_dummies.groupby(['Model','Company','Q_id'],as_index=False)['AP'].apply(lambda x: x.tolist()[-1]).to_frame()
```

```python
r1_new=basis_ranker.groupby(['Model'])['Model','Annoted_as_candidate'].mean()*100
r1_new=r1_new.reset_index()
r1_new=pd.DataFrame(r1_new.values,columns=['Model','MAP']).sort_values(by=['MAP'],ascending=False)

s1_new=basis_ranker.groupby(['Model','Company'])['Model','Company','Annoted_as_candidate'].mean()*100
s1_new=s1_new.reset_index()
s1_new=pd.DataFrame(s1_new.values,columns=['Model','Company','MAP'])

df0=r1_new
df0
```

### 6) Décompte des questions avec au moins une réponse (très pertinent)

```python
rnew=r1_dummies
by_question=rnew.groupby(['Model','Company','Q_id'],as_index=False)['Model','Annoted_as_candidate','Annotation_answer','Annotation_clue'].sum()
by_question['Total_Queries']=1
by_question['Annoted_as_candidate']=by_question['Annoted_as_candidate']>0
by_question['Annotation_answer']=by_question['Annotation_answer']>0
by_question['Annotation_clue']=by_question['Annotation_clue']>0
by_question=by_question.groupby(['Model'],as_index=False)['Annoted_as_candidate','Annotation_answer','Annotation_clue','Total_Queries'].sum()

```

```python
by_question=by_question[['Model','Annotation_answer','Annotation_clue','Annoted_as_candidate','Total_Queries']].sort_values(by='Annotation_answer',ascending=False)
```

```python
by_question["%Q_ANSWERED"]=np.round(by_question["Annotation_answer"]/by_question["Total_Queries"]*100,4)
by_question["%Q_C&A"]=np.round(by_question["Annoted_as_candidate"]/by_question["Total_Queries"]*100,4)
df6=by_question[['Model','%Q_ANSWERED','%Q_C&A']]
df6

```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="whitegrid")

# Initialize the matplotlib figure
f, ax = plt.subplots(figsize=(12, 2.5))

# Plot the total 
sns.set_color_codes("pastel")
sns.barplot(x="Total_Queries", y="Model", data=by_question,
            label="Not answered", color="palevioletred")

# Plot the answers and clue count
sns.set_color_codes("pastel")
sns.barplot(x="Annoted_as_candidate", y="Model", data=by_question,
            label="Clue Answer", color="navy")


sns.set_color_codes("muted")
sns.barplot(x="Annotation_answer", y="Model", data=by_question,
            label="True Answer", color="b")

# Add a legend and informative axis label
ax.legend(ncol=2, loc="lower right", frameon=True)
ax.set(xlim=(0, 117), ylabel="",
       xlabel="Question answered ESG Risk Assesemnt")
sns.despine(left=True, bottom=True)
```
